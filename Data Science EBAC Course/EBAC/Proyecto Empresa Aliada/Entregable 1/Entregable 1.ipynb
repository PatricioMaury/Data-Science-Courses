### Cargar tablas de datos en Pandas

```python
# Utiliza la función de Pandas para cargar cada tabla de datos desde el archivo csv o xlsx en un DataFrame.
# (DIM_CATEGORY, DIM_PRODUCT, DIM_SEGMENT, DIM_CALENDAR, FACT_SALES).
import pandas as pd
# Cargar archivos CSV
df_category = pd.read_csv('DIM_CATEGORY.csv')
df_sales = pd.read_csv('FACT_SALES.csv')

# Cargar archivos Excel
df_product = pd.read_excel('DIM_PRODUCT.xlsx')
df_segment = pd.read_excel('DIM_SEGMENT.xlsx')
df_calendar = pd.read_excel('DIM_CALENDAR.xlsx')
```

### Realizar la limpieza de datos

```python
# Corrige posibles inconsistencias en los datos, como errores tipográficos o formatos incorrectos.
# Identifica y maneja valores nulos en cada DataFrame (si es que los hay).
# Identifica si hay duplicados de los DataFrames y en caso de que los haya, eliminalos.

# DIM_CATEGORY
df_category['CATEGORY'] = df_category['CATEGORY'].str.strip().str.upper()
df_category = df_category.dropna()
df_category = df_category.drop_duplicates()

# DIM_PRODUCT
df_product['BRAND'] = df_product['BRAND'].str.strip().str.upper()
df_product['ITEM_DESCRIPTION'] = df_product['ITEM_DESCRIPTION'].str.strip().str.upper()
df_product = df_product.dropna()
df_product = df_product.drop_duplicates()

# DIM_SEGMENT
df_segment['SEGMENT'] = df_segment['SEGMENT'].str.strip().str.upper()
df_segment = df_segment.dropna()
df_segment = df_segment.drop_duplicates()

# DIM_CALENDAR
df_calendar['DATE'] = pd.to_datetime(df_calendar['DATE'], errors='coerce')
df_calendar = df_calendar.dropna()
df_calendar = df_calendar.drop_duplicates()

# FACT_SALES
df_sales['TOTAL_UNIT_SALES'] = pd.to_numeric(df_sales['TOTAL_UNIT_SALES'], errors='coerce')
df_sales['TOTAL_VALUE_SALES'] = pd.to_numeric(df_sales['TOTAL_VALUE_SALES'], errors='coerce')
df_sales = df_sales.dropna()
df_sales = df_sales.drop_duplicates()
```

### Unir los DataFrames relevantes

```python
# Unión de FACT_SALES con DIM_PRODUCT
df_ventas_productos = pd.merge(df_sales, df_product, how='left', left_on='ITEM_CODE', right_on='ITEM')
df_ventas_productos
```

```python
# Convertir ambas columnas CATEGORY a tipo texto para que coincidan
df_ventas_productos['CATEGORY'] = df_ventas_productos['CATEGORY'].astype(str)
df_category['CATEGORY'] = df_category['CATEGORY'].astype(str)
# Unión con DIM_CATEGORY
df_ventas_productos_categorias = pd.merge(df_ventas_productos, df_category, how='left', on='CATEGORY')
df_ventas_productos_categorias
```

```python
# Unión con DIM_SEGMENT
# Convertir columnas a string en ambos DataFrames para asegurar compatibilidad
columnas_union = ['CATEGORY', 'FORMAT', 'ATTR1', 'ATTR2', 'ATTR3']
for col in columnas_union:
    df_ventas_productos_categorias[col] = df_ventas_productos_categorias[col].astype(str)
    df_segment[col] = df_segment[col].astype(str)

# Hacer la unión final
df_final = pd.merge(df_ventas_productos_categorias, df_segment, how='left', on=columnas_union)
df_final
```

### Aplicar transformaciones necesarias

```python
# Estandariza los formatos de las columnas, como las fechas o categorías, para asegurar la consistencia
# en todo el conjunto de datos.
# Realiza cualquier transformación adicional requerida para preparar los datos para el análisis,
# como la creación de nuevas columnas calculadas o la agrupación de datos.

# Convertir columnas de texto clave a mayúsculas y eliminar espacios
columnas_texto = ['CATEGORY', 'BRAND', 'FORMAT', 'ATTR1', 'ATTR2', 'ATTR3', 'SEGMENT']
for col in columnas_texto:
    if col in df_final.columns:
        df_final[col] = df_final[col].str.strip().str.upper()

# Convertir columnas numéricas a tipo numérico (por si hay errores de formato)
df_final['TOTAL_UNIT_SALES'] = pd.to_numeric(df_final['TOTAL_UNIT_SALES'], errors='coerce')
df_final['TOTAL_VALUE_SALES'] = pd.to_numeric(df_final['TOTAL_VALUE_SALES'], errors='coerce')
```

```python
# Crear columna de precio promedio por unidad
df_final['UNIT_PRICE'] = df_final['TOTAL_VALUE_SALES'] / df_final['TOTAL_UNIT_SALES']
df_final
```

```python

# Agrupar ventas por categoría
ventas_por_categoria = df_final.groupby('CATEGORY')[['TOTAL_UNIT_SALES', 'TOTAL_VALUE_SALES']].sum().reset_index()
ventas_por_categoria
```

```python
# Creo que sería buena idea también incluir el df_calendar
# Asegurar que 'WEEK' sea del mismo tipo en ambos DataFrames
df_final['WEEK'] = df_final['WEEK'].astype(int)
df_calendar['WEEK'] = df_calendar['WEEK'].astype(int)

# Unir por la columna 'WEEK'
df_final = pd.merge(df_final, df_calendar, how='left', on='WEEK')
```

### Guardar el conjunto de datos consolidado

```python
# Guarda el DataFrame consolidado en un nuevo archivo CSV o Excel para su uso posterior en el análisis.
df_final.to_excel('DATOS_CONSOLIDADOS.xlsx', index=False)
```
