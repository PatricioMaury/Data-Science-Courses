```python
conteo = dict()
nombres = ['Juan', 'Pedro', 'Juan', 'Ramiro']
# Para cada nombre en la lista de nombres
# Checar si el nombre ya existe en el diccionario "conteo"
# Si el nombre no está en la lista, inicializar frecuencia en 1
# En caso contrario (ya está en la lista), incrementar su frecuencia en 1
for nombre in nombres:
    if nombre not in conteo:
        conteo[nombre] = 1
    else:
        conteo[nombre] = conteo[nombre] + 1
conteo
```

```python
archivo = open("Clase 2.4 Trabalenguas.txt")
archivo
```

```python
for texto in archivo:
    palabras = texto.split()
palabras
```

```python
conteo = dict()
for nombre in palabras:
    if nombre not in conteo:
        conteo[nombre] = 1
    else:
        conteo[nombre] = conteo[nombre] + 1
conteo
```
