```python
print(set([5, 2, 5, 1, 1.5])) #Inicializar con lista
print(set((5, 2, 5, 1, 1.5))) #Inicializar con tupla
print(set("52511.5")) #Inicializar con string
```

```python
conjunto2 = set([5, 3, 5, 6])
conjunto3 = set([4, 2])
print(conjunto, conjunto2, conjunto3)
```

```python
print(conjunto.intersection(conjunto2)) #Elementos que aparecen en ambos conjuntos
```

```python
print(conjunto2.issubset(conjunto)) #Ver si los elementos del Conjunto 2 se incluyen en el 1
```

```python
print(conjunto3.issubset(conjunto)) #Ver si los elementos del Conjunto 3 se incluyen en el 1
```

```python
# Diccionarios
```

```python
diccionario = {1: "Uno", 2: "Dos"}
print(diccionario)
```

```python
diccionario[3] = "Tres"
print(diccionario)
```

```python
print(diccionario[2])
```

```python
dict_tipos = {1: "integer", 2.2: "float", "texto": "string", (1, 2): "tupla"} #Se puede usar cualquier tipo de dato, pero la key debe de ser inmutable
print(dict_tipos)
```

```python
dict_repeticion = {1: "Primero", 1: "Ultimo"} #Si la key se repite, toma el valor de la última repetición
print(dict_repeticion)
```

```python
print(diccionario, diccionario.keys(), diccionario.values(), diccionario.items())
```

```python
claves = diccionario.values()
print(claves)
```

```python
diccionario[1] = "One"
print(claves)
```

```python
diccionario.pop(2) #Eliminar una key(clave) con su valor
print(diccionario)
print(claves)
```
