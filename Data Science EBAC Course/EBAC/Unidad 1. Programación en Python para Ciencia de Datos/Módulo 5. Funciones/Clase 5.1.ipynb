```python
def saludo():
    print('Hola')

saludo()
```

```python
def saludo_nombre(nombre):
    print('Hola ' + nombre)

saludo_nombre('Juan')
```

```python
#Calcular área de un círculo
def area_circulo(radio):
    area = 3.1416 * radio ** 2
    return area

mi_area = area_circulo(2)
print(mi_area)
```

```python
def producto(x1, x2):
    return x1 * x2

producto(2, 4)
```

```python
#Función con valores predeterminados
def suma(a, b = 3):
    return a + b

resultado = suma(2)
print(resultado)
suma(2, 4)
```

```python
def duplica(numero):
    x = numero * 2
    return x
print(duplica(10))
```

```python
def mi_funcion(*amigos):
    print('Mi mejor amigo es ' + amigos[2])

mi_funcion('Pedro', 'José', 'Linus', 'Juan')
```

```python
def chicos(chico1, chico2, chico3):
    print('El chico más joven es ' + chico3)

chicos(chico3 = 'Emil', chico1 = 'Tobias', chico2 = 'Armando')
```

```python
def apellido(**chico):
    print('El apellido es ' + chico['apellido'])

apellido(npila = 'Juan', apellido = 'Suárez', edad = 15)
```

```python
def pais(country = 'México'):
    print('Soy de ' + country)

pais('Argentina')
pais()
```

```python
#Pase de listas en funciones
def lista_comida(comida):
    for x in comida:
        print(x)

frutas = ['manzana', 'fresa', 'sandía']
lista_comida(frutas)
```
