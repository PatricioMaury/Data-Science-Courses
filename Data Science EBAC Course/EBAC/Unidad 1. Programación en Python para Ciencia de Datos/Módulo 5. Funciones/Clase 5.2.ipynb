# Funciones

```python
def tabla_del(numero):
    resultados = []
    for i in range(1, 11):
        resultados.append(numero * i)
    return resultados
```

```python
tabla_del(5)
```

```python
tabla_del(7)
```

```python
for valor in range(1,11):
    print(tabla_del(valor))
```

```python
# Comprobar si una string es un palíndromo
def palindromo(string):
    izquierda = 0
    derecha = len(string) - 1
    while derecha >= izquierda:
        if not (string[izquierda] == string[derecha]):
            return False
        izquierda += 1
        derecha -= 1
    return True
```

```python
print(palindromo('rayar'))
print(palindromo('raya'))
print(palindromo('Rayar'))
```

```python
# Hacer que acepte mayúsculas
def palindromo2(string):
    izquierda = 0
    derecha = len(string) - 1
    while derecha >= izquierda:
        if not (string[izquierda].lower() == string[derecha].lower()):
            return False
        izquierda += 1
        derecha -= 1
    return True
    
print(palindromo2('Rayar'))
```

```python
# Función Tradicional
def cuadrado(x):
    return x ** 2
print(cuadrado(2))
```

```python
# Función Lambda
cuadradolambda = lambda x: x ** 2
print(cuadradolambda(2))
```

```python
suma = lambda x, y: x + y
suma(2, 3)
```

```python
(lambda x, y: x + y)(2, 3)
```

```python
def calcula(n):
    return lambda x: x * n
resultado = calcula(2)
print('El doble del número 15 es: ', resultado(15))
```

```python
inicia_con = lambda texto, x: True if x.startswith(texto) else False
print(inicia_con('P', 'Python'))
print(inicia_con('R', 'Python'))
```
