```python
# Desarrolle un programa en Python que a través de una función denominada invierte_cadena sea capaz de invertir una cadena de caracteres. 
# Por ejemplo, si ejecutamos: invierte_cadena("123abc") debemos obtener: "cba321".
```

```python
def invierte_cadena(cadena):
    cadena_invertida = ''
    for caracter in cadena:
        cadena_invertida = cadena_invertida + caracter
    return cadena_invertida
```

```python
invierte_cadena('123abc')
```

```python
# Escribe una función en Python llamada prueba_primo que tome un número como parámetro y compruebe si el número es primo o no.
# Nota: Un número primo es un número natural mayor que 1 y que no tiene divisores positivos más que 1 y él mismo.
#Por ejemplo, si ejecutamos: prueba_primo(7) debemos obtener: True.
```

```python
def prueba_primo(numero):
    if numero <= 1:
        return False
    elif numero == 2:
        return True  # 2 es el único número par primo
    elif numero % 2 == 0:
        return False  # Si es par y mayor que 2, no es primo

    for i in range(3, numero, 2):  # Comprobamos solo números impares
        if numero % i == 0:
            return False

    return True
```

```python
prueba_primo(7)
```

```python
prueba_primo(31990)
```

```python
prueba_primo(31991)
```
