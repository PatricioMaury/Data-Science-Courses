```python
n = int(input('Capture el valor de la n: ')) #Si N es 3, la suma sería 1+2+3 = 6
suma = 0
i = 1
while i <= n:
    suma += i
    i += 1

print('La suma es: ', suma)
```

```python
#Determinar números pares e impares del 1 al 9

numeros = (1, 2, 3, 4, 5, 6, 7, 8, 9) #Tupla
valores = len(numeros)
impares = 0
pares = 0

contador = 1
while contador <= valores:
    if numeros[contador - 1] % 2 == 0:
        pares += 1
    else:
        impares += 1
    contador += 1
print('Número de valores pares: ', pares)
print('Número de valores impares: ', impares)
```

```python
#Secuencia de Fibonacci de 0 a 50
x, y = 0, 1
print(x)
print(y)

while x + y < 50:
    x = x + y
    print(x)
    y = x + y
    print(y)
```

```python
#Alternativa avanzada
x, y = 0, 1
print(x)

while y < 50:
    print(y)
    x, y = y, x + y
```
