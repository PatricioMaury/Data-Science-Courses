# FOR

```python
numeros = [6, 5, 3, 8, 4, 2, 5, 4, 11]
suma = 0
for valor in numeros:
    suma = suma + valor
print('La suma es: ', suma)
```

```python
numeros = [6, 5, 3, 8, 4, 2, 5, 4, 11]
suma = 0
for valor in numeros:
    suma = suma + valor
print(f'La suma es: {suma}')
```

```python
frutas = ['manzana', 'fresa', 'cereza']
for fruta in frutas:
    print(fruta)
```

```python
for letra in 'plátano':
    print(letra)
```

```python
total = 0
for i in 5, 7, 11, 13:
    print(i)
    total += i
print('Total: ', total)
```

```python
for x in range(6):
    print(x)
```

```python
for x in range(2, 10):
    print(x)
```

```python
for x in range(2, 10, 3):
    print(x)
```

```python
for x in range(6):
    print(x)
else:
    print('Proceso terminado')
```
