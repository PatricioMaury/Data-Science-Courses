# WHILE

```python
n = 4
while n > 0:
    print(n)
    n = n - 1
print('Fin')
```

```python
n = 1
while n > 1:
    print('Uno')
    print('Dos')
print('Fin')
```

```python
while True:
    linea = input('Escribe un texto: ')
    if linea == 'Fin':
        break
    print(linea)
print('Proceso terminado')
```

```python
while True:
    linea = input('Escribe un texto: ')
    if linea[0] == '#':
        continue
    if linea == 'Fin':
        break
    print(linea)
print('Proceso terminado')
```

```python
i = 1
while i < 6:
    print(i)
    i += 1
else:
    print('i ya no es mayor a 6')
```
