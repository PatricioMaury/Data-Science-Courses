```python
#Desarrolle un programa en Python que mediante el uso de ciclos calcule el factorial del número que indique el usuario.
#En caso de que el valor proporcionado sea inválido (por ejemplo, que sea negativo o fraccionado), deberá devolver un mensaje de "Error de captura".
#Pruebe su código calculando los factoriales de 4, 5 y 6 respectivamente así como el de 7.5.
#Almacena el resultado en la varible {factorial}.
#El factorial de un número entero positivo n, representado por n!, es la multiplicación de n por sus antecesores mayores o iguales a 1.
#Por ejemplo: 4! = 4x3x2x1 = 24
```

```python
cadena = input("Dame un entero positivo: ")

# Puedes usar la estructura "try/except" que aprendimos en el módulo 2.
# comience su código aquí:
try:
    n = int(cadena)
    if n < 0:
        raise ValueError("¡Error en el valor ingresado! ¡Debe ser un entero positivo!")
except ValueError:
    print("Error de captura")
else:
    factorial = 1
    # Si n es válido, calculamos el factorial usando un ciclo
    for i in range(1, n + 1):
        factorial = factorial * i
    print(f"El factorial de {n} es: {factorial}")
```

```python
#Desarrolle un programa en Python que mediante el uso de un ciclo while determine si
#un número n (proporcionado por el usuario) puede ser expresado como 2^k donde K es entero.
#El valor de n debe ser un entero entre 2 y 100000.
#Si n = 8, puede ser expresado como 2^3 = 8, por tanto, el resultado es True.
#Si n = 32, puede ser expresado como 2^5 = 32, por tanto, el resultado es True.
#Si n = 40, no puede ser expresado como 2^K = 40, por tanto, el resulto es False.
#El resultado debe ser un booleano almacenado en el variable es_potencia_de_2.
```

```python
try:
	n = int(input("Ingrese el número: "))
	if n < 2:
		raise ValueError("¡Error en el valor ingresado! ¡Debe ser un número entero entre 2 y 100000!") from None
except ValueError as e:
	raise ValueError("¡Error en el valor ingresado! ¡Debe ser un número entero entre 2 y 100000!") from None

es_potencia_de_2 = False

# comience su código aquí:
while n % 2 == 0:
    n = n // 2  # Divide n entre 2
if n == 1:
    es_potencia_de_2 = True

print(es_potencia_de_2)
```
