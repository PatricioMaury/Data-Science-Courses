```python
tablas = int(input('Número de tablas a calcular: '))

for i in range(1, tablas + 1):
    print('Tabla de multiplicar del: ', i)
    contador = 1
    while contador < 11:
        print(i * contador)
        contador += 1
    print('\n')
```

```python
tablas = int(input('Número de tablas a calcular: '))

for i in range(1, tablas + 1):
    print('Tabla de multiplicar del: ', i)
    contador = 1
    while contador < 11:
        print(i * contador, end = ' ')
        contador += 1
    print('\n')
```
