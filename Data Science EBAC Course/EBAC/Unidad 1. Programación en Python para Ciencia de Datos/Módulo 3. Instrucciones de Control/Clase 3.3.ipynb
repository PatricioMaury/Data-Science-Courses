# IF, ELSE, ELIF

```python
x = 4
if x > 2:
    print('Más grande')
else:
    print('Más pequeño')
print('Fin')
```

```python
x = float(input('Dame un número: '))
if x < 2:
    print('x = ', x)
    print('Pequeño')
elif x < 10:
    print('x = ', x)
    print('Mediano')
else:
    print('x = ', x)
    print('Grande')
print('\nFin')
```

```python
x = float(input('Dame un número: ')) #Hacerlo más sencillo pero con más elif
print('x = ', x)
if x < 2:
    print('Pequeño')
elif x < 10:
    print('Mediano')
elif x < 20:
    print('Grande')
else:
    print('Gigante')
print('\nFin')
```

```python
x = float(input('Dame un número: ')) #Puedes poner cuantos elif como quieras, pero solo un if y else alineados
print('x = ', x)
if x < 2:
    print('Pequeño')
elif x < 10:
    print('Mediano')
elif x < 20:
    print('Grande')
elif x < 40:
    print('Muy grande')
elif x < 100:
    print('Enorme')
else:
    print('Gigante')
print('\nFin')
```
