# Condicionales (IF)

```python
a = 33
b = 200
if b > a:
    print('b es mayor que a')
```

```python
if 'f' in 'foo':
    print('1')
    print('2')
    print('3')
```

```python
if 'f' in 'foo': print('1'); print('2'); print('3') #Alternativa a ejecución en varios renglones
```

```python
i = 15
if i > 11: print('i es mayor que 11')
```

```python
a = 200
b = 33
c = 500
if a > b and c > a:
    print('Ambas condiciones son verdaderas')
```

```python
a = 200
b = 33
c = 500
if a > b or c > a:
    print('Al menos una condición es verdadera')
```

```python
x = 5
if x < 10:
    print('Pequeño')
if x > 20:
    print('Grande')
print('Fin')
```

```python
x = 5
if x == 5:
    print('Igual a 5')
if x > 4:
    print('Mayor que 4')
if x >= 5:
    print('Mayor o igual que 5')
if x < 6: print('Menor que 6')
if x <= 5: print('Menor o igual que 5')
if x != 6: print('Diferente de 6')
```

### Decisiones Anidadas (NESTED)

```python
x = input('Dame un número: ')
if float(x) > 1:
    print('Mayor a 1')
    if float(x) < 100:
        print('Menor a 100')
print('Proceso terminado')
```
