## Ejemplo de uso de super()

```python
class Figura:
    def __init__(self, nombre):
        self.nombre = nombre

    def area(self):
        pass

class Circulo(Figura):
    def __init__(self, radio):
        super().__init__("Círculo")
        self.radio = radio

    def area(self):
        return 3.14159 * self.radio * self.radio

class Triangulo(Figura):
    def __init__(self, base, altura):
        super().__init__("Triángulo")
        self.base = base
        self.altura = altura

    def area(self):
        return 0.5 * self.base * self.altura

# Uso de las clases
circulo = Circulo(5)
triangulo = Triangulo(4, 3)

print(f"Área del {circulo.nombre}: {circulo.area()}")
print(f"Área del {triangulo.nombre}: {triangulo.area()}")
    
```
