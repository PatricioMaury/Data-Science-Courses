```python
# Escriba una clase en Python denominada 'Conversion' que permita la conversión de un número romano a un número entero a través de
# un método llamado 'romano_entero'. Por ejemplo, si definimos un objeto de esta clase llamado 'numero', al invocar: 
# 'numero.romano_entero('MMMCMLXXXVI')', debemos obtener: 3986.
```

```python
class Conversion:
    """Clase para convertir números romanos a enteros"""
    
    def __init__(self):
        self.valores = {
            'I': 1, 'V': 5, 'X': 10, 'L': 50,
            'C': 100, 'D': 500, 'M': 1000
        }
    
    def romano_entero(self, romano):
        """Convierte un número romano en su equivalente entero"""
        entero = 0
        prev = 0
        
        for letra in reversed(romano):
            valor = self.valores[letra]
            if valor < prev:
                entero -= valor
            else:
                entero += valor
            prev = valor
        
        return entero

# Ejemplo de uso
numero = Conversion()
print(numero.romano_entero('MMMCMLXXXVI'))  # Salida: 3986
```

```python
# Método 2
class Conversion(object):
    """Clase para convertir números romanos a enteros"""

    def __init__(self):
        self.valores = {
            'I': 1, 'V': 5, 'X': 10, 'L': 50,
            'C': 100, 'D': 500, 'M': 1000
        }

    def romano_entero(self, romano):
        """Convierte un número romano en su equivalente entero"""
        entero = 0
        longitud = len(romano)

        for i in range(longitud):
            valor_actual = self.valores[romano[i]]
            
            # Verificar si hay un número mayor a la derecha (sustracción)
            if i < longitud - 1 and valor_actual < self.valores[romano[i + 1]]:
                entero -= valor_actual
            else:
                entero += valor_actual

        return entero

# Ejemplo de uso
numero = Conversion()
print(numero.romano_entero('MMMCMLXXXVI'))  # Salida: 3986
```
