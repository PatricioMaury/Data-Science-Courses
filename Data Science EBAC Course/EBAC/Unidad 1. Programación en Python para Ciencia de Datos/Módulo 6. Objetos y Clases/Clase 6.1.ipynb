# Clases y Objetos

```python
# Creación de una nueva clase
class Mi_clase:
    pass
```

```python
# Creacion de un nuevo objeto
objeto = Mi_clase
objeto.x = 5
objeto.y = 10

print(objeto.x)
print(objeto.y)
```

```python
# Creación de una nueva clase con atributos iniciales
class Mi_clase:
    x = 5
    y = 10
```

```python
objeto = Mi_clase()
print(objeto.x)
print(objeto.y)
```

```python
# Definición de un método para una clase
class Aviso:
    def saludar(self):
        print('Hola, buen día')
```

```python
texto = Aviso()
texto.saludar()
```

```python
# Definición de una clase para almacenar los datos de estudiantes
class Estudiante:
    def aprobado(self):
        if self.calif >= 60:
            return True
        else:
            return False
```

```python
estudiante1 = Estudiante()
estudiante1.nombre = 'Juan'
estudiante1.calif = 80
aprobo1 = estudiante1.aprobado()
print(aprobo1)
```

```python
estudiante2 = Estudiante()
estudiante2.nombre = 'Aracely'
estudiante2.calif = 50
aprobo2 = estudiante2.aprobado()
print(aprobo2)
```

```python
# Manejo de un constructor
class Estudiante:
    def aprobado(self):
        if self.calif >= 60:
            return True
        else:
            return False

    def __init__(self, nombre, calif):
        self.nombre = nombre
        self.calif = calif
```

```python
estudiante3 = Estudiante('José', 90)
print(estudiante3.nombre)
print(estudiante3.calif)
```

```python
aprobo3 = estudiante3.aprobado()
print(aprobo3)
```
