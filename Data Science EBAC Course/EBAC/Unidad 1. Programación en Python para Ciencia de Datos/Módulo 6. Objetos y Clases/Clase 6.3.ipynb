## Tipos de Constructores

```python
# Constructor Predeterminado
class Expertos:
    def __init__(self):
        self.texto = 'Sesión experta'

    def imprime_texto(self):
        print(self.texto)

elemento = Expertos()
elemento.imprime_texto()
```

```python
# Constructor Parametrizado
class Adicion:
    respuesta = 0

    def __init__(self, primero, segundo):
        self.primero = primero
        self.segundo = segundo

    def calcula(self):
        self.respuesta = self.primero + self.segundo

    def mostrar(self):
        print('Primer número = ' + str(self.primero))
        print('Segundo número = ' + str(self.segundo))
        print('Adición de dos números = ' + str(self.respuesta))
```

```python
obj = Adicion(1000, 2000)
obj.calcula()
obj.mostrar()
```

## Herencias

```python
# Herencia Simple
class Estudiante(object):                # Clase 'Padre o Madre'
    def __init__(self, edad, nombre):
        self.edad = edad
        self.nombre = nombre

class Derecho(Estudiante):              # Clase 'Hija'
    def presentarse(self):
        print(f'Soy {self.nombre} tengo {self.edad} años y estudio Derecho')

Manuel = Derecho (26, 'Manuel Gómez')
Manuel.presentarse()
```

```python
# Herencia Múltiple
class Estudiante(object):                # Clase 'Padre o Madre' no. 1
    def __init__(self, edad, nombre):
        self.edad = edad
        self.nombre = nombre

class Instituto(object):                # Clase 'Padre o Madre' no. 2
    def presentarinstituto(self):
        print('Estudio en el Instituto de Leyes')

class Derecho(Estudiante, Instituto):   # Clase 'Hija'
    def presentarse(self):
        print(f'Soy {self.nombre} tengo {self.edad} años y estudio Derecho')

Manuel = Derecho(26, 'Manuel Contreras')
Manuel.presentarse()
Manuel.presentarinstituto()
```

```python
# Defina una clase padre denominada 'Figura'. Desarrolle dos clases hijas; 'Círculo' y Triángulo'.
# Deben ser capaces de calcular el área en cada caso a partir de los parámetros requeridos por cada figura.
# Para el círculo se requiere el radio y para el triángulo se requiere base y altura.

class Figura(object):
    def area(self):
        pass

class Circulo(Figura):
    """Objeto de tipo círculo"""
    def __init__(self, radio=0):
        self.radio = radio

    def area(self):
        return 3.1416 * (self.radio ** 2)

class Triangulo(Figura):
    '''Objeto de tipo triángulo'''
    def __init__(self, base=0, altura=0):
        self.base = base
        self.altura = altura

    def area(self):
        return self.base * self.altura / 2
```

```python
c = Circulo(4)
c.area()
```

```python
t = Triangulo(3, 5)
t.area()
```
